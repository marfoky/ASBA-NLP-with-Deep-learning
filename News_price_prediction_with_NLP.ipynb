```python
# Merge the two datasets on the 'Date' column
reported_figures = pd.merge(non_farm_payroll, unemployment_rate, on='Expected Impact / Date', how='inner')
reported_figures
```

```python
# Sort the combined dataset by the 'Date' column
reported_figures = reported_figures.sort_values(by='Expected Impact / Date')
# Convert the 'Expected Impact / Date' column to datetime
reported_figures['Expected Impact / Date'] = pd.to_datetime(reported_figures['Expected Impact / Date'], errors='coerce')
# Remove 'K' and convert numerical columns to integers
for col in ['Actual_x', 'Forecast_x', 'Previous_x']:
    reported_figures[col] = reported_figures[col].str.replace('K', '').astype(float) * 1000
# Remove '%' and convert percentage columns to floats
for col in ['Actual_y', 'Forecast_y', 'Previous_y']:
    reported_figures[col] = reported_figures[col].str.replace('%', '').astype(float) / 100
# View the results
reported_figures
```

```python
# Connect to the SQLite database
conn = sqlite3.connect('news_text.db')
# Query the sqlite_master table to get the list of tables
query = "SELECT name FROM sqlite_master WHERE type='table';"
# Execute the query and fetch all results
cursor = conn.cursor()
cursor.execute(query)
# tables = cursor.fetchall()
query = "SELECT * FROM employment_situation"
# load the data into a DataFrame
news_data = pd.read_sql_query(query, conn)
news_data
# Close the connection after loading the data
conn.close()
```

***Merging both structured and unstructured news data***

```python
# Convert date columns to datetime format if not already done
news_data['updated_date'] = pd.to_datetime(news_data['updated_date'])
reported_figures['Expected Impact / Date'] = pd.to_datetime(reported_figures['Expected Impact / Date'])
# Find the latest starting date to ensure both DataFrames have overlapping dates
start_date = max(news_data['updated_date'].min(), reported_figures['Expected Impact / Date'].min())
# Filter both DataFrames to start from the same date
news_data_filtered = news_data[news_data['updated_date'] >= start_date]
reported_figures_filtered = reported_figures[reported_figures['Expected Impact / Date'] >= start_date]
# Sort both DataFrames by date
news_data_filtered = news_data_filtered.sort_values(by='updated_date')
reported_figures_filtered = reported_figures_filtered.sort_values(by='Expected Impact / Date')
# Merge the filtered and sorted DataFrames
merged_df = pd.merge(
    reported_figures_filtered, news_data_filtered,
    left_on='Expected Impact / Date',
    right_on='updated_date',
    how='outer'
)
# Sort the final merged DataFrame by date
merged_df = merged_df.sort_values(by='Expected Impact / Date')
merged_df
```

```python
import pandas as pd
import re

# Load the dataset (assuming it's already loaded as a DataFrame named `df`)
# Adjust the file path if loading from a CSV or other sources.
# df = pd.read_csv('your_dataset.csv')  # Uncomment if loading from a CSV

# Extract the `text_column`
texts = merged_df['text_column'].dropna()

# Function to clean and preprocess text
def preprocess_text(text):
    # Remove newline characters
    text = text.replace('\n', ' ')
    # Remove special characters, numbers, and multiple spaces
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

# Apply preprocessing to the text column
merged_df['cleaned_text'] = texts.apply(preprocess_text)

# Display a sample of the cleaned text
merged_df[['text_column', 'cleaned_text']].head()
```

```python
from transformers import BertTokenizer, TFBertForSequenceClassification
from transformers import pipeline
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# Load FinBERT pre-trained model and tokenizer
model_name = "yiyanghkust/finbert-tone"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = TFBertForSequenceClassification.from_pretrained(model_name)

# Define a sentiment analysis function
def analyze_sentiment(texts):
    # Tokenize the texts
    inputs = tokenizer(texts, return_tensors="tf", padding=True, truncation=True, max_length=512)
    
    # Run the inputs through the FinBERT model
    outputs = model(inputs['input_ids'], attention_mask=inputs['attention_mask'])
    logits = outputs.logits
    
    # Apply softmax to calculate probabilities
    probs = tf.nn.softmax(logits, axis=-1)
    
    # Map sentiment categories
    sentiment_labels = ["neutral", "positive", "negative"]
    sentiments = []
    for prob in probs:
        sentiment_index = tf.argmax(prob).numpy()
        sentiment_score = prob[sentiment_index].numpy()
        sentiment = {
            "label": sentiment_labels[sentiment_index],
            "confidence": sentiment_score
        }
        sentiments.append(sentiment)
    
    return sentiments

# Apply sentiment analysis to the cleaned text
merged_df['sentiment_analysis'] = merged_df['cleaned_text'].apply(lambda x: analyze_sentiment([x])[0])

# Extract sentiment labels and confidence scores into separate columns
merged_df['sentiment_label'] = merged_df['sentiment_analysis'].apply(lambda x: x['label'])
merged_df['sentiment_confidence'] = merged_df['sentiment_analysis'].apply(lambda x: x['confidence'])

# Display the results
merged_df[['cleaned_text', 'sentiment_label', 'sentiment_confidence']].head()
```
