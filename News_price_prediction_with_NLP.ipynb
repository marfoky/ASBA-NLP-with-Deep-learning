***Data cleaning***

```python
# Loading the datasets
price_df = pd.read_csv('price_data')
non_farm_payroll = pd.read_csv('non_farm_payroll.csv', encoding='iso-8859-1')
unemployment_rate = pd.read_csv('US_unemployment_rate.csv', encoding='iso-8859-1')
non_farm_payroll
```

```python
unemployment_rate
```

```python
price_df
```

```python
# Merge the two datasets on the 'Date' column
combined_data = pd.merge(non_farm_payroll, unemployment_rate, on='Expected Impact / Date', how='inner')
combined_data
```

```python
# Convert the 'Date' column to datetime format
combined_data['Expected Impact / Date'] = pd.to_datetime(combined_data['Expected Impact / Date'], errors='coerce')
# Sort the combined dataset by the 'Date' column
combined_data_sorted = combined_data.sort_values(by='Expected Impact / Date')
# Reset the index (optional, but it keeps the DataFrame clean after sorting)
# combined_data_sorted = combined_data_sorted.reset_index(drop=True)
combined_data_sorted
```
